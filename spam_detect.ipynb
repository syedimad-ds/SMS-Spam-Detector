```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import make_pipeline
from sklearn.metrics import accuracy_score


df = pd.read_csv('/content/spam.csv', encoding='latin-1')[['v1', 'v2']]
df.columns = ['label', 'message']


X_train, X_test, y_train, y_test = train_test_split(df['message'], df['label'], test_size=0.2, random_state=42)


model = make_pipeline(CountVectorizer(), MultinomialNB())
model.fit(X_train, y_train)


print(f"Accuracy: {accuracy_score(y_test, model.predict(X_test))*100:.2f}%")


def predict_spam(message):

    return model.predict([message])[0]


my_msg = "Congratulations! You've won a $1000 Walmart gift card. Go to http://bit.ly/123 to claim now."
print(f"Message: '{my_msg}' -> Prediction: {predict_spam(my_msg)}")
```

```python
import joblib
joblib.dump(model, 'spam_model.pkl')

print("Model saved successfully!")
```
